```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


house_price = pd.read_csv('train.csv')
house_price.head()
```

```python
house_price.shape
```

```python
house_price.isnull().sum()
```

```python
plt.hist(house_price['SalePrice'], bins=20)
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
corr_matrix = house_price.corr()
corr_matrix['SalePrice'].sort_values(ascending=False)
```

```python
import seaborn as sns

sns.heatmap(corr_matrix, cmap='coolwarm')
plt.show()
```

```python
plt.hist(house_price['LotArea'], bins=20)
plt.xlabel('Lot Area')
plt.ylabel('Frequency')
plt.show()

sns.boxplot(x='MSZoning', y='SalePrice', data=house_price)
plt.show()
```

```python
# 9.Analyze the relationship between variables
# Analyze the relationship between variables using scatter plots or pair plots.

plt.scatter(house_price['GrLivArea'], house_price['SalePrice'])
plt.xlabel('GrLivArea')
plt.ylabel('Sale Price')
plt.show()

sns.pairplot(house_price[['LotArea', 'OverallQual', 'YearBuilt', 'SalePrice']])
plt.show()
```
